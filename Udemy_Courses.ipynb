```python
df['subject'].unique()
```

#2.knowing how many courses are (free) in every subject in the dataset


```python
df[df['is_paid'] == False].groupby('subject')['subject'].value_counts()
```

#3.knowing how many courses are(not free) in every subject in the dataset


```python
df[df['is_paid'] == True].groupby('subject')['subject'].value_counts()
```

#4.Find out the most subscribed course

```python
df.sort_values('num_subscribers',ascending = False)[['course_title','num_subscribers','is_paid','num_lectures','subject']].head()
```

#5.Find out the course with the least subscription

```python
df.sort_values('num_subscribers',ascending = True)[['course_title','num_subscribers','is_paid','num_lectures','subject']].head()
```

#6.Find out the course with biggest number of lectures

```python
df.sort_values('num_lectures',ascending = False)[['course_title','num_subscribers','is_paid','num_lectures','subject']].head()
```
